```python


for col in df.columns:
    missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing*100)))
```

```python
# when I looked at the data set. I saw that except one column all columns are object. 
# I need to convert(object to float) couple of columns. These columns are price, number_of_reviews,
# number_available_in_stock and average_review_rating.
```

```python
def mapprice(v):
    if pd.isnull(v):
        return 0
    try:
        return float(v[1:])
    except ValueError:
        return 0
    
df['price'] = df['price'].map(mapprice)
```

```python
df['number_of_reviews'] = df['number_of_reviews'].str.replace(',', '').astype(float)
```

```python
def mapnumber_available_in_stock(v):
    if pd.isnull(v):
        return 0 
    try:
        vv = v.replace('new', '')
        return float(vv)
    except ValueError:
        return 0
df['number_available_in_stock'] = df['number_available_in_stock'].map(mapnumber_available_in_stock)
```

```python
def mapaverage_review_rating(v):
    if pd.isnull(v):
        return 0
    try:
        vv = v.replace('out of 5 stars', '')
        return float(vv)
    except ValueError:
        return 0 
df['average_review_rating'] = df['average_review_rating'].map(mapaverage_review_rating)
```
